```python
def summarize_profitable_exit(df):
    # Group the data by relevant fields
    summary = df.groupby(['Day of Week', 'Option Type', 'entry_time', 'moneyness', 'Strike Distance', 'DTE']).agg(
        total_trades=('contract_symbol', 'count'),  # Count total trades
        success_5pct=('5%', 'mean'),  # Percentage of 5% profit success
        success_10pct=('10%', 'mean'),  # Percentage of 10% profit success
        success_15pct=('15%', 'mean'),  # Percentage of 15% profit success
        success_20pct=('20%', 'mean'),  # Percentage of 20% profit success
        success_25pct=('25%', 'mean'),  # Percentage of 25% profit success
        success_30pct=('30%', 'mean'),  # Percentage of 30% profit success
        success_35pct=('35%', 'mean'),  # Percentage of 35% profit success
        success_40pct=('40%', 'mean'),  # Percentage of 40% profit success
        success_45pct=('45%', 'mean'),  # Percentage of 45% profit success
        success_50pct=('50%', 'mean')  # Percentage of 50% profit success
    ).reset_index()

    # Convert the success rates from proportions to percentages
    summary['success_5pct'] = summary['success_5pct'] * 100
    summary['success_10pct'] = summary['success_10pct'] * 100
    summary['success_15pct'] = summary['success_15pct'] * 100
    summary['success_20pct'] = summary['success_20pct'] * 100
    summary['success_25pct'] = summary['success_25pct'] * 100
    summary['success_30pct'] = summary['success_30pct'] * 100
    summary['success_35pct'] = summary['success_35pct'] * 100
    summary['success_40pct'] = summary['success_40pct'] * 100
    summary['success_45pct'] = summary['success_45pct'] * 100
    summary['success_50pct'] = summary['success_50pct'] * 100

    # Filter out rows where no profit targets were met
    # summary = summary[(summary['success_5pct'] > 0) | (summary['success_10pct'] > 0) | (summary['success_15pct'] > 0)
    #                   | (summary['success_20pct'] > 0) | (summary['success_25pct'] > 0) | (summary['success_30pct'] > 0)
    #                   | (summary['success_35pct'] > 0) | (summary['success_40pct'] > 0) | (summary['success_45pct'] > 0)
    #                   | (summary['success_50pct'] > 0)]

    return summary
```

```python
# Call the summarization function with your result_df
summary_df = summarize_profitable_exit(result_df)

# Display the summarized results
print(summary_df.head())
```

## Analyze the summaries / find alpha

```python
def rank_top_achievers(df):
    # Calculate composite score based on success rates
    df['composite_score'] = (0.1 * df['success_5pct'] + 
                             0.2 * df['success_10pct'] + 
                             0.3 * df['success_15pct'] + 
                             0.4 * df['success_20pct'] +
                             0.5 * df['success_25pct'] +
                             0.6 * df['success_30pct'] +
                             0.7 * df['success_35pct'] +
                             0.8 * df['success_40pct'] +
                             0.9 * df['success_45pct'] +
                             1.0 * df['success_50pct']
                             )

    # Rank the results by composite score
    df_ranked = df.sort_values(by='composite_score', ascending=False).reset_index(drop=True)

    return df_ranked

# Call the function with your summarized dataframe (from the previous step)
ranked_df = rank_top_achievers(summary_df)

# Display the top achievers
print(ranked_df.head(10))

```

```python
def rank_top_achievers(df, min_trades=5):
    # Filter out combinations that don't meet the minimum trade threshold
    df_filtered = df[df['total_trades'] >= min_trades]

    # Calculate composite score based on success rates
    df_filtered['composite_score'] = (0.1 * df['success_5pct'] + 
                                      0.2 * df['success_10pct'] + 
                                      0.3 * df['success_15pct'] + 
                                      0.4 * df['success_20pct'] +
                                      0.5 * df['success_25pct'] +
                                      0.6 * df['success_30pct'] +
                                      0.7 * df['success_35pct'] +
                                      0.8 * df['success_40pct'] +
                                      0.9 * df['success_45pct'] +
                                      1.0 * df['success_50pct']
                                      )

    # Rank the results by composite score
    df_ranked = df_filtered.sort_values(by='composite_score', ascending=False).reset_index(drop=True)

    return df_ranked

# Example usage with a threshold of at least 5 trades:
ranked_df = rank_top_achievers(summary_df, min_trades=12)

# Display the top achievers
print(ranked_df.head(10))
```
